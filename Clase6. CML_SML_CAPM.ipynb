# Equilibrio de Mercado

El equilibrio de mercado se logra cuando la oferta y la demanda de un activo se igualan, determinando así su precio en un momento específico. Este equilibrio es dinámico y puede cambiar debido a diversos factores que afectan las condiciones del mercado.

Del lado de las acciones e instrumentos financieros podemos verlo como la convergencia de la demanda de la acción y venta de la misma, dando un **precio justo o a mercado**.   

**Ejemplo:**  
Consideremos una acción X que cotiza en $70 dólares en este momento, ¿cómo se elegió ese precio? Esto se da, con la compra venta de la misma, hay una cierta posición que trata de comprar la acción en cierto precio (demanda) y existe la contraparte que trata de vender la acción en otro precio, esto da lugar a una convergencia de un precio justo en el mercado. 

En terminos formales, podemos ver este caso como:  
**Precio Bid (oferta):**
El precio bid es el precio al cual un comprador está dispuesto a adquirir un activo. Es el precio máximo que alguien está dispuesto a pagar por un activo en ese momento. En otras palabras, si estás buscando vender un activo, el precio bid representa la mejor oferta disponible para comprar ese activo en ese momento.  

**Precio Ask (demanda):**  
El precio ask es el precio al cual un vendedor está dispuesto a vender un activo. Es el precio mínimo que alguien está dispuesto a aceptar por un activo en ese momento. Si estás buscando comprar un activo, el precio ask representa la mejor oferta disponible para adquirir ese activo en ese momento.  

La diferencia entre el precio bid y el precio ask se conoce como el **"spread"**. Un spread más estrecho generalmente indica una mayor liquidez en el mercado, mientras que un spread más amplio puede indicar menor liquidez. Los traders y inversores utilizan estos precios bid y ask para realizar transacciones y evaluar la actividad del mercado.

# Capital Asset Pricing Model (CAPM) para Valoración de Activos Financieros:

El Modelo de Valoración de Activos de Capital (CAPM) es una teoría financiera desarrollada en la década de 1960 que busca relacionar el riesgo y el rendimiento de los activos financieros. Fue propuesto de manera independiente por William F. Sharpe, John Lintner y Jan Mossin.

El CAPM tuvo un impacto significativo en la teoría financiera y la práctica de la inversión. Introdujo la idea de que el riesgo diversificable puede reducirse a través de la construcción de carteras bien diversificadas, y que el riesgo sistemático es el factor que realmente afecta el rendimiento de un activo. Aunque el CAPM ha sido criticado y refinado con el tiempo debido a sus suposiciones simplificadas y limitaciones, sigue siendo una herramienta fundamental en la valoración de activos y en la toma de decisiones de inversión.

## Tipos de Riesgos  
1. **Riesgo Sistemático (o Riesgo de Mercado)**:
El riesgo sistemático se refiere al riesgo que está asociado con el mercado en su conjunto y que no puede eliminarse a través de la diversificación. Este tipo de riesgo afecta a todos los activos en el mercado y es causado por factores macroeconómicos y eventos que afectan a todas las empresas e industrias. Algunos ejemplos de factores que pueden generar riesgo sistemático incluyen cambios en las tasas de interés, fluctuaciones económicas, eventos políticos y condiciones generales del mercado.
Dado que el riesgo sistemático es inherente al mercado en su conjunto, no se puede eliminar simplemente diversificando la cartera. En otras palabras, no importa cuántos activos se tengan en una cartera, siempre habrá cierto nivel de riesgo sistemático.

2. **Riesgo No Sistemático (o Riesgo Específico)**:
El riesgo no sistemático es el riesgo que es específico de una empresa o activo en particular y que puede diversificarse. Se refiere a los riesgos únicos que afectan a una empresa o a un sector particular de la economía, pero que no se propagan a través de todo el mercado. Ejemplos de riesgo no sistemático pueden incluir problemas de gestión, eventos legales específicos de la empresa o cambios en la demanda de productos de una empresa.
Dado que el riesgo no sistemático se refiere a factores específicos de una empresa o industria, es posible reducir o eliminar este riesgo a través de la diversificación. Al construir una cartera diversificada que incluya activos de diferentes sectores y empresas, es posible mitigar el impacto de riesgos específicos de un solo activo.

## Supuestos de CAPM

**1. Mercados eficientes**: El CAPM parte del supuesto de que los mercados financieros son perfectamente eficientes, lo que significa que toda la información relevante está disponible para los inversionistas de manera instantánea y que los precios de los activos reflejan completamente esta información.

**2. Inversionistas racionales**: Se asume que los inversionistas son racionales y buscan maximizar su utilidad a través de decisiones de inversión óptimas. Esto implica que los inversionistas considerarán todos los datos disponibles y tomarán decisiones basadas en una evaluación precisa de riesgos y rendimientos.

**3. No hay impuestos ni costos de transacción**: El CAPM ignora los impuestos y los costos de transacción asociados con la compra y venta de activos. Esto permite simplificar los cálculos y la relación entre riesgo y rendimiento.

**4. Inversores con horizontes idénticos**: El modelo supone que todos los inversores tienen horizontes de inversión idénticos, lo que significa que todos tienen el mismo horizonte temporal y la misma capacidad para asumir riesgos.


**5. Los inversores pueden prestar y pedir prestado al mismo tipo de interés**: El modelo presume que los inversionistas pueden tomar prestado o prestar dinero al mismo tipo de interés sin restricciones.

**6. Diversificación perfecta:** El CAPM asume que los inversionistas pueden construir carteras perfectamente diversificadas, lo que significa que pueden mantener una combinación de activos que elimine completamente el riesgo no sistemático (diversificable).El CAPM asume que los activos pueden ser comprados o vendidos en cantidades infinitamente pequeñas, lo que facilita la construcción de carteras óptimas y la medición precisa del riesgo.

**7. Expectativas homogéneas:** Se supone que todos los inversionistas tienen las mismas expectativas respecto a las tasas de rendimiento esperadas y la distribución de riesgos.

## Fallas del modelo en la práctica
**Tal vez la falla más importante de este modelo es que el CAPM:**

1. **No toma en cuenta el riesgo no sitemático.**  
  
2. El CAMP asume que los inversionistas solo se preocupan de la desviación estandar **(volatilidad del activo)** y del retorno de los **activos (rendimientos)**.  


3. Que **los activos no se correlacionan entre sí y solo lo hacen con el mercado.** Al decir que los activos están correlacionados significa que el mercado de consumo impacta al mercado de materias primas. **El CAPM asume que no.**   


4. Otro supuesto es que todos los iversionistas están **valuando los activos en el mismo lapso de tiempo.**    


5. Asume que se pueden tomar **prestamos bancarios con interes = a la tasa libre de riesgo y que podemos invertir a la tasa libre de riesgo.** Esto el banco no le conviene porque al prestarte a ti cae en riesgo, sería más fácil que él lo invierta en el $r_f$.  

6. Que **los impuestos no nos afecta** en la compra venta de activos.  


7. Asume que todos **los inversionistas hacen la misma estimación de todos los parámetros del capm**, tiempo, retorno, volatilidad, correlación entre activos, y esto es completamente falso.  

**El CAPM se basa en la formula del sharpe ratio que es una formula para encontrar el portafolio de inversión que maximice el rendimiento por unidad de riesgo, el capm asume que todos los inversionistas lo logran, es decir que todos los inversionistas estamos invirtiendo en el portafolio óptimo.**

**En resumen, el riesgo sistemático afecta a todo el mercado y no puede eliminarse mediante la diversificación, mientras que el riesgo no sistemático es específico de un activo o empresa y puede diversificarse.**

El rendimiento esperado $(R_i)$ de un activo se calcula utilizando la siguiente fórmula del CAPM:

$$\bar{r_i} = r_f + \beta_i \times (E(r_m) - r_f)$$

Donde:
- $\bar{r_i}$: Rendimiento esperado del activo.
- $r_f$: Tasa de rendimiento libre de riesgo.
- $\beta_i$: Coeficiente beta del activo.
- $E(r_m)$: Rendimiento esperado del mercado.

El término $(E(r_m) - r_f)$ representa la prima de riesgo del mercado, que refleja la compensación adicional que los inversionistas requieren por asumir riesgo.

La beta del activo se calcula como:  
$$\beta_i = \frac{Cov(r_i,r_m)}{\sigma^2_m}$$

**NOTAS:**  
- La $\beta_i$ es el riesgo sistemático con respecto del mercado.   

- Cuando hablamos del rendimiento del mercado, estamos diciendo los retornos que nos entrega el índice que tomemos. 


## Ejemplo:  
Supongamos que:  
$r_f$ = 11%  
$r_m$ = 15%  
$\beta_i$ = 1.5  

```python
# Variables
rf = 0.11
rm = 0.15
beta = 1.5
```

```python
capm = rf + beta*(rm-rf)
print(capm)
```

Tenemos un CAPM de x, esto nos dice que nuestro activo debe tener por lo menos x% de rendimiento. Esto no implica que x% sea el rendimiento de nuestro activo, sino que es una referencia de cuánto debería ser el rendimiento.  
Si el activo rinde más que el CAPM tendremos una subvaluación del modelo, en otro caso tendremos una sobre valuación.

**¿Qué pasa con una beta de 0 y 1?**

```python
# Caso beta = 1
rf = 0.11
rm = 0.15
beta = 1
capm = rf + beta*(rm-rf)
print(capm)
```

El rendimiento esperado del activo nos está diciendo que el activo replica exactamente al mercado.

```python
# Caso beta = 0
rf = 0.11
rm = 0.15
beta = 0
capm = rf + beta*(rm-rf)
print(capm)
```

 Esto pasa porque el $r_f$ no incluye riesgo sitemático. Es decir, estamos invirtiendo solamento en $r_f$. Esto se le conoce como la **Security Market Line (SML) o la Línea de Mercado de Capitales.**

# Diferencia entre CML y SML  
**El CML traza sólo las carteras (más) eficientes, pero el SML traza todas las carteras (incluidas las carteras ineficientes).**  
Abajo a la izquierda: se trazan dos carteras ( puntos naranjas ). Ambos han esperado una rentabilidad del +16,5%, pero la cartera eficiente (en el CML) tiene una volatilidad menor (18,7% frente a 21,6%); es más eficiente. Todos los puntos de la CML son los más eficientes: todos tienen el mismo Sharpe más alto = 56,5% (pendiente de la CML). La cartera menos eficiente tiene un riesgo específico (también conocido como idiosincrásico) = 10,8%, mientras que la cartera eficiente tiene un riesgo específico cero. Podemos decir que el subóptimo está menos "bien diversificado".

Pero ambas carteras (a la izquierda) se ubican en la misma ubicación en el SML: E(rt) = 16,5% frente a beta = 1,60. Ambos tienen la misma beta, pero al SML no le importa la diferencia en el riesgo específico.
```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Datos
rf = 0.11  # Tasa de rendimiento libre de riesgo
rm = 0.15  # Rendimiento esperado del mercado

# Valores de beta
betas = np.arange(-1.5, 1.6, 0.1)

# Cálculo de los rendimientos esperados (CAPM)
capm_values = rf + betas * (rm - rf)

# Crear el gráfico
plt.figure(figsize=(8, 6)) # Tamaño (ancho, altura)
plt.plot(betas, capm_values, marker='o', linestyle='-', color='b')
plt.title('Modelo de Valoración de Activos de Capital (CAPM)')
plt.xlabel('Beta')
plt.ylabel('Rendimiento Esperado')
plt.grid(True, color='0.7', linestyle='--')
plt.axhline(y=rf, color='r', linestyle='--', label='Tasa libre de riesgo')
plt.axhline(y=rm, color='g', linestyle='--', label='Rendimiento del mercado')
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.8)
plt.legend()
plt.show()
```

**Observemos sobre las línea verde y roja que las distancias de manera horizontal es una medida de riesgo no sitemático.**

```python
# Datos
rf = 0.11  # Tasa de rendimiento libre de riesgo
rm = 0.15  # Rendimiento esperado del mercado

# Valores de beta
betas = np.arange(-1.5, 1.6, 0.1)

# Cálculo de los rendimientos esperados (CAPM)
capm_values = rf + betas * (rm - rf)

# Redondear las betas a 3 decimales
betas_rounded = np.round(betas, 3)

# Crear un DataFrame
data = {'CAPM': capm_values, 'Betas': betas_rounded}
df = pd.DataFrame(data)

# Mostrar el DataFrame
df
```

Las betas negativas significan que los activos dan rendimiento a la inversa del mercado, es decir que se mueven opuestos a el índice.

```python
# Por qué la beta de -1 no nos da -0.15
rf = 0.11
rm = 0.15
beta = -1
capm = rf + beta*(rm-rf)
print(capm)
```

**La beta que se dio en un principio ya está internamente usando los rendimientos del activo i, pero estos rendimientos son pasados, mientras que el rendimiento del activo por CAPM es el esperado en función al tiempo que estaremos tomando.**
